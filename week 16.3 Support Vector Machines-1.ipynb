**Q1. What is the mathematical formula for a linear SVM?**

**ANSWER:-------**


A linear Support Vector Machine (SVM) aims to find the optimal hyperplane that separates the data points of different classes with the maximum margin. The mathematical formulation of a linear SVM involves:

1. **Decision Function:**
   The decision function for a linear SVM can be written as:
   \[
   f(\mathbf{x}) = \mathbf{w} \cdot \mathbf{x} + b
   \]
   where:
   - \(\mathbf{x}\) is the input feature vector.
   - \(\mathbf{w}\) is the weight vector.
   - \(b\) is the bias term.
   - \(\cdot\) represents the dot product.

2. **Classification Rule:**
   A data point \(\mathbf{x}\) is classified based on the sign of \(f(\mathbf{x})\):
   \[
   y = \text{sign}(f(\mathbf{x})) = \text{sign}(\mathbf{w} \cdot \mathbf{x} + b)
   \]

3. **Optimization Problem:**
   To find the optimal hyperplane, we solve the following optimization problem:
   \[
   \min_{\mathbf{w}, b} \frac{1}{2} \|\mathbf{w}\|^2
   \]
   subject to the constraints:
   \[
   y_i (\mathbf{w} \cdot \mathbf{x}_i + b) \geq 1 \quad \forall i
   \]
   where:
   - \(y_i\) is the class label of the \(i\)-th training example (\(y_i \in \{-1, 1\}\)).
   - \(\mathbf{x}_i\) is the \(i\)-th training example.
   - \(\|\mathbf{w}\|^2\) is the squared Euclidean norm of the weight vector, which is equivalent to \(\mathbf{w} \cdot \mathbf{w}\).

4. **Lagrangian Dual Problem:**
   The problem can also be formulated in its dual form using Lagrange multipliers \(\alpha_i\):
   \[
   \max_{\alpha} \sum_{i=1}^{n} \alpha_i - \frac{1}{2} \sum_{i=1}^{n} \sum_{j=1}^{n} \alpha_i \alpha_j y_i y_j (\mathbf{x}_i \cdot \mathbf{x}_j)
   \]
   subject to:
   \[
   \sum_{i=1}^{n} \alpha_i y_i = 0 \quad \text{and} \quad \alpha_i \geq 0 \quad \forall i
   \]
   where \(n\) is the number of training examples.

5. **Weight Vector and Bias:**
   The weight vector \(\mathbf{w}\) and bias \(b\) can be recovered from the solution of the dual problem:
   \[
   \mathbf{w} = \sum_{i=1}^{n} \alpha_i y_i \mathbf{x}_i
   \]
   \[
   b = y_k - \mathbf{w} \cdot \mathbf{x}_k \quad \text{for any} \quad \alpha_k > 0
   \]

These equations describe the theoretical framework of a linear SVM and how it finds the optimal hyperplane to separate different classes with the maximum margin.

**Q2. What is the objective function of a linear SVM?**

**ANSWER:------**


The objective function of a linear Support Vector Machine (SVM) is designed to find the optimal hyperplane that maximizes the margin between the classes while minimizing classification errors. It combines two main components: the margin maximization term and the hinge loss term (for handling misclassifications).

### Hard Margin SVM (Linearly Separable Case)
For a linearly separable case (hard margin SVM), the objective is to maximize the margin between the two classes. This can be formulated as:

1. **Primal Formulation:**
   Minimize the norm of the weight vector \(\mathbf{w}\):
   \[
   \min_{\mathbf{w}, b} \frac{1}{2} \|\mathbf{w}\|^2
   \]
   subject to the constraints:
   \[
   y_i (\mathbf{w} \cdot \mathbf{x}_i + b) \geq 1 \quad \forall i
   \]
   where:
   - \(y_i \in \{-1, 1\}\) is the class label of the \(i\)-th training example.
   - \(\mathbf{x}_i\) is the \(i\)-th training example.
   - \(\mathbf{w}\) is the weight vector.
   - \(b\) is the bias term.

### Soft Margin SVM (Non-linearly Separable Case)
For the non-linearly separable case (soft margin SVM), slack variables \(\xi_i\) are introduced to allow some misclassifications. The objective function includes a penalty for misclassifications:

1. **Primal Formulation:**
   Minimize the norm of the weight vector \(\mathbf{w}\) and the hinge loss:
   \[
   \min_{\mathbf{w}, b, \xi} \frac{1}{2} \|\mathbf{w}\|^2 + C \sum_{i=1}^{n} \xi_i
   \]
   subject to the constraints:
   \[
   y_i (\mathbf{w} \cdot \mathbf{x}_i + b) \geq 1 - \xi_i \quad \forall i
   \]
   \[
   \xi_i \geq 0 \quad \forall i
   \]
   where:
   - \(\xi_i\) are the slack variables representing the degree of misclassification.
   - \(C\) is a regularization parameter that controls the trade-off between maximizing the margin and minimizing the classification error.

### Dual Formulation:
The dual formulation of the SVM optimization problem focuses on Lagrange multipliers \(\alpha_i\):

1. **Dual Formulation:**
   \[
   \max_{\alpha} \sum_{i=1}^{n} \alpha_i - \frac{1}{2} \sum_{i=1}^{n} \sum_{j=1}^{n} \alpha_i \alpha_j y_i y_j (\mathbf{x}_i \cdot \mathbf{x}_j)
   \]
   subject to:
   \[
   \sum_{i=1}^{n} \alpha_i y_i = 0 \quad \text{and} \quad 0 \leq \alpha_i \leq C \quad \forall i
   \]

### Summary:
The primary objective function of a linear SVM, in its primal form for the soft margin case, is:
\[
\min_{\mathbf{w}, b, \xi} \frac{1}{2} \|\mathbf{w}\|^2 + C \sum_{i=1}^{n} \xi_i
\]
subject to:
\[
y_i (\mathbf{w} \cdot \mathbf{x}_i + b) \geq 1 - \xi_i \quad \forall i
\]
\[
\xi_i \geq 0 \quad \forall i
\]

This formulation seeks to balance the maximization of the margin (first term) with the minimization of the classification errors (second term). The parameter \(C\) controls the trade-off between these two goals.

**Q3. What is the kernel trick in SVM?**

**ANSWER:-------**



The kernel trick in Support Vector Machines (SVMs) is a technique that allows SVMs to perform classification in higher-dimensional spaces without explicitly computing the coordinates of the data in that space. This is particularly useful for handling non-linear relationships in the data.

### Why Use the Kernel Trick?
- **Non-linear Separation**: In many real-world problems, the data cannot be linearly separated in the original feature space. By mapping the data into a higher-dimensional space, it becomes easier to find a linear separation.
- **Efficiency**: Directly computing the transformation to a high-dimensional space and then solving the optimization problem can be computationally expensive. The kernel trick allows this process to be done more efficiently.

### How the Kernel Trick Works
The kernel trick uses a kernel function \( K(\mathbf{x}_i, \mathbf{x}_j) \) that computes the dot product of the data points in the higher-dimensional space without explicitly mapping them to that space. This allows SVMs to operate in the original input space while implicitly performing computations in a higher-dimensional feature space.

### Common Kernel Functions
1. **Linear Kernel**:
   \[
   K(\mathbf{x}_i, \mathbf{x}_j) = \mathbf{x}_i \cdot \mathbf{x}_j
   \]
   This is equivalent to the standard dot product in the input space and does not perform any mapping.

2. **Polynomial Kernel**:
   \[
   K(\mathbf{x}_i, \mathbf{x}_j) = (\mathbf{x}_i \cdot \mathbf{x}_j + c)^d
   \]
   where \( c \) is a constant and \( d \) is the degree of the polynomial.

3. **Radial Basis Function (RBF) Kernel**:
   \[
   K(\mathbf{x}_i, \mathbf{x}_j) = \exp\left(-\gamma \|\mathbf{x}_i - \mathbf{x}_j\|^2\right)
   \]
   where \( \gamma \) is a parameter that defines the width of the Gaussian function. This kernel maps data into an infinite-dimensional space.

4. **Sigmoid Kernel**:
   \[
   K(\mathbf{x}_i, \mathbf{x}_j) = \tanh(\alpha \mathbf{x}_i \cdot \mathbf{x}_j + c)
   \]
   where \( \alpha \) and \( c \) are kernel parameters. This is similar to the activation function in neural networks.

### Mathematical Formulation
Given a kernel function \( K(\mathbf{x}_i, \mathbf{x}_j) \), the dual form of the SVM optimization problem can be written as:
\[
\max_{\alpha} \sum_{i=1}^{n} \alpha_i - \frac{1}{2} \sum_{i=1}^{n} \sum_{j=1}^{n} \alpha_i \alpha_j y_i y_j K(\mathbf{x}_i, \mathbf{x}_j)
\]
subject to:
\[
\sum_{i=1}^{n} \alpha_i y_i = 0 \quad \text{and} \quad 0 \leq \alpha_i \leq C \quad \forall i
\]

The decision function for making predictions is:
\[
f(\mathbf{x}) = \sum_{i=1}^{n} \alpha_i y_i K(\mathbf{x}_i, \mathbf{x}) + b
\]

### Benefits of the Kernel Trick
- **Flexibility**: Allows SVMs to handle a wide variety of data structures and distributions.
- **Scalability**: Enables working with high-dimensional spaces without the computational burden of explicitly transforming the data.
- **Performance**: Often results in better classification performance for non-linearly separable data.

The kernel trick is a powerful tool that extends the applicability of SVMs to a broader range of problems by enabling efficient computation in high-dimensional spaces.

**Q4. What is the role of support vectors in SVM Explain with example**

**ANSWER:-------**


### Role of Support Vectors in SVM

Support Vectors play a crucial role in the functioning of Support Vector Machines (SVMs). They are the data points that lie closest to the decision boundary (or hyperplane) and are most critical in determining the position and orientation of this hyperplane. The SVM model is essentially built around these support vectors, and the optimal hyperplane is found by maximizing the margin between the support vectors of the two classes.

### Key Points about Support Vectors:
1. **Critical for Model Training**: Only the support vectors are used to determine the optimal hyperplane. The rest of the data points do not influence the hyperplane directly.
2. **Margin Maximization**: The hyperplane is positioned in such a way that it maximizes the distance (margin) from the closest support vectors of each class.
3. **Robustness**: SVMs are robust to outliers because the position of the hyperplane is influenced only by the support vectors, not by other data points.
4. **Sparse Solution**: Because only a subset of the training data (the support vectors) is used in the final model, SVMs can be computationally efficient in practice.

### Example
Consider a simple 2D classification problem where we have two classes of data points: Class A (blue points) and Class B (red points).

#### Step-by-Step Example:
1. **Data Visualization**:
   Imagine we plot the data points on a 2D plane. We see that the two classes can be separated by a line.

2. **Finding the Hyperplane**:
   The SVM algorithm identifies the optimal hyperplane that separates the two classes with the maximum margin. This hyperplane is represented as:
   \[
   \mathbf{w} \cdot \mathbf{x} + b = 0
   \]
   where \(\mathbf{w}\) is the weight vector and \(b\) is the bias term.

3. **Identifying Support Vectors**:
   The points that lie on the edge of the margin (i.e., the points closest to the hyperplane) are the support vectors. These points satisfy:
   \[
   \mathbf{w} \cdot \mathbf{x}_i + b = \pm 1
   \]
   There will be at least one support vector for each class. These points are critical as they determine the margin width and the position of the hyperplane.

4. **Formulation**:
   The objective is to maximize the margin, which is equivalent to minimizing \(\frac{1}{2} \|\mathbf{w}\|^2\) subject to the constraint that all points are correctly classified with a margin of at least 1. This can be written as:
   \[
   \min_{\mathbf{w}, b} \frac{1}{2} \|\mathbf{w}\|^2
   \]
   subject to:
   \[
   y_i (\mathbf{w} \cdot \mathbf{x}_i + b) \geq 1 \quad \forall i
   \]

#### Visualization:
Below is a conceptual visualization of a 2D SVM classification:

```
Class A (Blue):  o
Class B (Red):   x

Support Vectors: circled points

Hyperplane: Solid line
Margins: Dashed lines

     +---------+---------+
     |         |         |
     |   o     |   x     |
     |         |         |
     |---------|---------| <-- Margin boundary for Class A
     |         |         |
     |    o    |    x    | <-- Support vectors (circled)
     |         |         |
     |---------+---------| <-- Hyperplane
     |         |         |
     |    x    |    o    | <-- Support vectors (circled)
     |         |         |
     |---------|---------| <-- Margin boundary for Class B
     |         |         |
     |    x    |    o    |
     |         |         |
     +---------+---------+
```

In this example:
- The solid line represents the optimal hyperplane.
- The dashed lines represent the margins on either side of the hyperplane.
- The circled points are the support vectors, which are crucial for determining the hyperplane.

### Summary
Support vectors are the backbone of SVMs, defining the decision boundary by maximizing the margin between classes. By relying on these critical points, SVMs achieve a robust and effective separation of data points, which can be visualized and appreciated through examples and visual aids.

**Q5. Illustrate with examples and graphs of Hyperplane, Marginal plane, Soft margin and Hard margin in

**ANSWER:--------**


### Hyperplane, Marginal Plane, Soft Margin, and Hard Margin in SVM

Let's illustrate these concepts with examples and graphs.

### 1. Hyperplane
A hyperplane in an SVM is the decision boundary that separates different classes. In a 2D space, it is a line; in a 3D space, it is a plane; in higher dimensions, it is a hyperplane.

#### Example:
Consider a 2D space with two classes of points (blue circles and red crosses):

```
Class A (Blue):  o
Class B (Red):   x

     o   o   o   x   x
     o       |       x
     o       |       x
     o       |       x
     o       o   x   x
```

In this example, the hyperplane (solid vertical line) separates the two classes.

### 2. Marginal Plane
Marginal planes are the boundaries that define the margin around the hyperplane. They are parallel to the hyperplane and are determined by the support vectors.

#### Example:
Using the same 2D example, the hyperplane and marginal planes are illustrated below:

```
Class A (Blue):  o
Class B (Red):   x

      Margin boundary for Class A
      |
     o   o   o   x   x
     o   ---(Hyperplane)---   x
     o       |       x
     o       |       x
     o       o   x   x
     |
      Margin boundary for Class B
```

In this graph:
- The hyperplane is the solid vertical line.
- The marginal planes are the dashed vertical lines, which are parallel to the hyperplane.

### 3. Hard Margin
A hard margin SVM is used when the data is linearly separable with no errors. It finds the maximum margin hyperplane that perfectly separates the classes without any misclassifications.

#### Example:
```
Class A (Blue):  o
Class B (Red):   x

      Margin boundary for Class A
      |
     o   o   o   x   x
     o   ---(Hyperplane)---   x
     o       |       x
     o       |       x
     o       o   x   x
     |
      Margin boundary for Class B
```

### 4. Soft Margin
A soft margin SVM allows for some misclassifications. It introduces slack variables to handle cases where the data is not perfectly separable, aiming to balance the margin size and the classification error.

#### Example:
```
Class A (Blue):  o
Class B (Red):   x

      Margin boundary for Class A
      |
     o   o   o   x   x
     o   ---(Hyperplane)---   x
     o       |       x
     o       |       x
     o       o   x   x   o   <-- Misclassified point
     |
      Margin boundary for Class B
```

In this graph:
- The hyperplane and marginal planes are the same as in the hard margin example.
- The point circled in red on the margin boundary for Class B represents a misclassified point allowed by the soft margin.


````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm

# Create data
np.random.seed(1)
X = np.r_[np.random.randn(20, 2) - [2, 2], np.random.randn(20, 2) + [2, 2]]
Y = [0] * 20 + [1] * 20

# Fit SVM with a linear kernel
clf = svm.SVC(kernel='linear', C=1000)
clf.fit(X, Y)

# Get the separating hyperplane
w = clf.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(-5, 5)
yy = a * xx - (clf.intercept_[0]) / w[1]

# Plot the points
plt.scatter(X[:, 0], X[:, 1], c=Y, s=30, cmap=plt.cm.Paired)

# Plot the separating hyperplane
plt.plot(xx, yy, 'k-')

# Plot the margins
margin = 1 / np.sqrt(np.sum(clf.coef_ ** 2))
yy_down = yy - np.sqrt(1 + a ** 2) * margin
yy_up = yy + np.sqrt(1 + a ** 2) * margin
plt.plot(xx, yy_down, 'k--')
plt.plot(xx, yy_up, 'k--')

# Highlight support vectors
plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], facecolors='none', edgecolors='k', s=100, linewidth=1.5)
plt.title('SVM with Hard Margin')
plt.show()
````

````python
#SOFT MARGIN
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm

# Create data
np.random.seed(1)
X = np.r_[np.random.randn(20, 2) - [2, 2], np.random.randn(20, 2) + [2, 2]]
Y = [0] * 20 + [1] * 20

# Introduce some noise for soft margin
X[::5] += 5 * np.random.rand(8, 2)

# Fit SVM with a linear kernel
clf = svm.SVC(kernel='linear', C=1)
clf.fit(X, Y)

# Get the separating hyperplane
w = clf.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(-5, 10)
yy = a * xx - (clf.intercept_[0]) / w[1]

# Plot the points
plt.scatter(X[:, 0], X[:, 1], c=Y, s=30, cmap=plt.cm.Paired)

# Plot the separating hyperplane
plt.plot(xx, yy, 'k-')

# Plot the margins
margin = 1 / np.sqrt(np.sum(clf.coef_ ** 2))
yy_down = yy - np.sqrt(1 + a ** 2) * margin
yy_up = yy + np.sqrt(1 + a ** 2) * margin
plt.plot(xx, yy_down, 'k--')
plt.plot(xx, yy_up, 'k--')

# Highlight support vectors
plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], facecolors='none', edgecolors='k', s=100, linewidth=1.5)
plt.title('SVM with Soft Margin')
plt.show()
````

**Q6. SVM Implementation through Iris dataset.**

~ Load the iris dataset from the scikit-learn library and split it into a training set and a testing set

~ Train a linear SVM classifier on the training set and predict the labels for the testing set

~ Compute the accuracy of the model on the testing set

~ Plot the decision boundaries of the trained model using two of the features

~ Try different values of the regularisation parameter C and see how it affects the performance of
the model.

````python
# Step 1: Load the Iris Dataset and Split into Training and Testing Sets


from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import numpy as np

# Load the iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
````

````python
#Step 2: Train a Linear SVM Classifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Train a linear SVM classifier
svc = SVC(kernel='linear', C=1.0)
svc.fit(X_train, y_train)

# Predict the labels for the testing set
y_pred = svc.predict(X_test)

# Compute the accuracy of the model on the testing set
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
````

````python
##Step 3: Plot the Decision Boundaries
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# Select two features for visualization
X_train_2d = X_train[:, :2]
X_test_2d = X_test[:, :2]

# Train the model again using only two features
svc_2d = SVC(kernel='linear', C=1.0)
svc_2d.fit(X_train_2d, y_train)

# Create a mesh to plot the decision boundaries
x_min, x_max = X_train_2d[:, 0].min() - 1, X_train_2d[:, 0].max() + 1
y_min, y_max = X_train_2d[:, 1].min() - 1, X_train_2d[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

# Predict the class for each point in the mesh
Z = svc_2d.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision boundaries
plt.contourf(xx, yy, Z, alpha=0.8, cmap=ListedColormap(('red', 'green', 'blue')))

# Plot the training points
plt.scatter(X_train_2d[:, 0], X_train_2d[:, 1], c=y_train, edgecolors='k', cmap=ListedColormap(('red', 'green', 'blue')))
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('SVM Decision Boundaries with Linear Kernel')
plt.show()
````

````python
#Step 4: Try Different Values of the Regularization Parameter C
Cs = [0.01, 0.1, 1, 10, 100]
accuracies = []

for C in Cs:
    svc = SVC(kernel='linear', C=C)
    svc.fit(X_train, y_train)
    y_pred = svc.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)
    print(f'C={C}: Accuracy={accuracy:.2f}')

# Plot accuracy vs C
plt.plot(Cs, accuracies, marker='o')
plt.xscale('log')
plt.xlabel('Regularization parameter C')
plt.ylabel('Accuracy')
plt.title('Effect of C on SVM Performance')
plt.show()
````

Implement a linear SVM classifier from scratch using Python and compare its
performance with the scikit-learn implementation.

**ANSWER:------**


To implement a linear SVM classifier from scratch and compare it with the scikit-learn implementation, we can follow these steps:

1. **Implement a linear SVM classifier from scratch.**
2. **Train the custom SVM on the Iris dataset.**
3. **Compare its performance with the scikit-learn implementation.**



````python
import numpy as np

class CustomSVM:
    def __init__(self, learning_rate=0.001, lambda_param=0.01, n_iters=1000):
        self.learning_rate = learning_rate
        self.lambda_param = lambda_param
        self.n_iters = n_iters
        self.w = None
        self.b = None

    def fit(self, X, y):
        n_samples, n_features = X.shape
        
        # Initialize weights and bias
        self.w = np.zeros(n_features)
        self.b = 0
        
        # Convert labels to -1 and 1
        y_ = np.where(y <= 0, -1, 1)
        
        for _ in range(self.n_iters):
            for idx, x_i in enumerate(X):
                condition = y_[idx] * (np.dot(x_i, self.w) - self.b) >= 1
                if condition:
                    self.w -= self.learning_rate * (2 * self.lambda_param * self.w)
                else:
                    self.w -= self.learning_rate * (2 * self.lambda_param * self.w - np.dot(x_i, y_[idx]))
                    self.b -= self.learning_rate * y_[idx]

    def predict(self, X):
        approx = np.dot(X, self.w) - self.b
        return np.sign(approx)
````

````python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Load the iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# We will use only two classes (binary classification)
X = X[y != 2]
y = y[y != 2]

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardize the dataset
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Train custom SVM
custom_svm = CustomSVM()
custom_svm.fit(X_train, y_train)
y_pred_custom = custom_svm.predict(X_test)

# Train scikit-learn SVM
svc = SVC(kernel='linear', C=1.0)
svc.fit(X_train, y_train)
y_pred_sklearn = svc.predict(X_test)

# Compute accuracy
accuracy_custom = accuracy_score(y_test, y_pred_custom)
accuracy_sklearn = accuracy_score(y_test, y_pred_sklearn)

print(f'Custom SVM Accuracy: {accuracy_custom:.2f}')
print(f'Scikit-learn SVM Accuracy: {accuracy_sklearn:.2f}')
````
